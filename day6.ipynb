```python
def parse_input(input_file):
    grid = []
    with open(input_file) as f:
        for line in f:
            line = line.rstrip()
            grid.append(line.split())
    nums = grid[:-1]
    nums = [[int(x) for x in row] for row in nums]
    ops = grid[-1]
    return nums, ops

def part1(input_file):
    nums, ops = parse_input(input_file)
    ans = 0
    for j in range(len(ops)):
        if ops[j] == '+':
            v = 0
        else:
            v = 1
        for i in range(len(nums)):
            if ops[j] == '+':
                v += nums[i][j]
            elif ops[j] == '*':
                v *= nums[i][j]  
        ans += v
    return ans
    
```

```python
def parse_input2(input_file):
    grid = []
    with open(input_file) as f:
        for line in f:
            grid.append([c for c in line if c != '\n'])
    return grid

def part2(input_file):
    grid = parse_input2(input_file)
    m, n = len(grid), len(grid[0])
    ans = 0
    right = n - 1
    while right > 0:
        left = right -1
        while left >= 0 and grid[m-1][left] == ' ':
            left -= 1
        op = grid[m-1][left]
        if op == '+':
             v = 0
        else:
             v = 1

        for j in range(right, left-1, -1):
            digits = [grid[i][j] for i in range(m-1)]
            num = int(''.join(digits).strip())
            if op == '+':
                v += num
            elif op == '*':
                v *= num
        ans += v
        right = left - 2
        
    return ans
```

```python
part2('input/day6_test.txt')
```
