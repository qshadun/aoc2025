```python
def load_input(input_file):
    ans = []
    with open(input_file) as f:
        for line in f:
            ans.append([c for c in line.rstrip()])
    return ans

directions = [(x, y) for x in (-1, 0, 1) for y in (-1, 0, 1) if not (x == 0 and y == 0)]

def count(grid, i, j):
    cnt = 0
    for di, dj in directions:
        ni, nj = i + di, j + dj
        if 0 <= ni < len(grid) and 0 <= nj < len(grid[0]):
            if grid[ni][nj] == '@':
                cnt += 1
    return cnt

def part1(input_file):
    grid = load_input(input_file)
    ans = 0
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] == '@':
                if count(grid, i, j) < 4:
                    ans += 1
    return ans
```

```python
def find_removable(grid):
    removable = []
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] == '@':
                if count(grid, i, j) < 4:
                    removable.append((i, j))
    return removable

def part2(input_file):
    grid = load_input(input_file)
    ans = 0
    while True:
        removable = find_removable(grid)
        if not removable:
            break
        for i, j in removable:
            grid[i][j] = '.'
        ans += len(removable)
    return ans
```
