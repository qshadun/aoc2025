```python
import re

def parse_input(input_file):
    ans = []
    with open(input_file) as f:
        for line in f:
            ans.append([int(c) for c in line.strip()])    
    return ans

def get_max(nums):
    suffix = [0]
    for n in reversed(nums):
        suffix.append(max(suffix[-1], n))
    suffix.reverse()
    ans = 0
    for i in range(len(nums) - 1):
        cur = nums[i] * 10 + suffix[i + 1]
        ans = max(ans, cur)
    return ans

def part1(input_file):
    rows = parse_input(input_file)
    ans = 0
    for row in rows:
        mv = get_max(row)
        ans += mv 
    return ans
```

```python
from functools import cache

def get_max2(nums):
    @cache
    def dp(i, remain):
        if i == len(nums):
            if remain > 0:
                return float('-inf')
            return 0
        if remain == 0:
            return 0
        
        take = dp(i + 1, remain - 1) + nums[i] * 10 ** (remain - 1)
        notake = dp(i + 1, remain)
        return max(take, notake)
    return dp(0, 12)

def part2(input_file):
    rows = parse_input(input_file)
    ans = 0
    for row in rows:
        mv = get_max2(row)
        # print(mv)
        ans += mv
    return ans
```
