```python
from collections import defaultdict

def parse_input(input_file):
    fresh, avail = [], []
    is_avail = False
    with open(input_file) as f:
        for line in f:
            line = line.rstrip()
            if not line:
                is_avail = True
                continue
            if not is_avail:
                low, up = line.split('-')
                fresh.append((int(low), int(up)))
            else:
                avail.append(int(line))
    return fresh, avail

def part1(input_file):
    fresh ,avail = parse_input(input_file)
    ans = 0
    for a in avail:
        for low, up in fresh:
            if low <= a <= up:
                ans += 1
                break
    return ans

def part2(input_file):
    fresh ,_ = parse_input(input_file)
    ans = 0
    fresh.sort()
    cur_low, cur_up = fresh[0]
    for low, up in fresh[1:]:
        if low <= cur_up + 1:
            cur_up = max(cur_up, up)
        else:
            ans += cur_up - cur_low + 1
            cur_low, cur_up = low, up
    ans += cur_up - cur_low + 1 
    return ans
```

```python
part1('input/day5_test.txt')
```
