```python
from functools import cache
from collections import deque

deltas = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def parse_input(input_file):
    shapes = []
    trees = []
    lines = []
    with open(input_file) as f:
        for line in f:
            if line.strip():
                lines.append(line.strip())
    
    i = 0
    while i < len(lines):
        if len(lines[i]) == 2:
            new_shape =[]
            for j in range(1, 4):
                new_shape.append([c for c in lines[i + j]])
            shapes.append(new_shape)
            i += 4
        else:
            xs = lines[i].split(' ')
            long, width = xs[0][:-1].split('x')
            trees.append((int(long), int(width), [int(x) for x in xs[1:]]))
            i += 1
    return shapes, trees


def part1(input_file):
    shapes, trees = parse_input(input_file)
    shape_size = []
    for shape in shapes:
        x = 0
        for row in shape:
            for c in row:
                if c == '#':
                    x += 1
        shape_size.append(x)
    
    ans = 0
    for w, h, ls in trees:
        if w * h >= sum(sz * x for sz, x in zip(shape_size, ls)):
            ans += 1
    return ans
```
