```python
import re

def part1(input_file):
    ans = 0
    pos = 50
    
    with open(input_file) as f:
        for line in f:
            dir = line[0]
            steps = int(line[1:])
            if dir == 'L':
                pos -= steps
            elif dir == 'R':
                pos += steps
            pos %= 100
            if pos == 0:
                ans += 1
    return ans
```

```python
part1('input/day1.txt')
```

```python
def part2(input_file):
    ans = 0
    pos = 50
    
    with open(input_file) as f:
        for line in f:
            dir = line[0]
            steps = int(line[1:])
            if steps > 100:
                ans += steps // 100
            steps %= 100
            if dir == 'L':
                if pos > 0 and steps > pos:
                    ans += 1
                    
                pos -= steps
            elif dir == 'R':
                if pos >0 and steps > (100 - pos):
                    ans += 1
                    
                pos += steps
            pos %=100
            if pos == 0:
                ans += 1
    return ans
```

```python
part2('input/day1_test.txt')
```
