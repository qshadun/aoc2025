```python
def parse(input_file):
    ans = []
    with open(input_file) as f:
        for line in f:
            for s in line.split(','):
                n1, n2 = s.split('-')
                ans.append((int(n1), int(n2)))
    return ans

def part1(input_file):
    ranges = parse(input_file)
    ans = 0
    for low, up in ranges:
        for x in range(low, up + 1):
            xs = str(x)
            if len(xs) % 2 == 0:
                mid = len(xs) // 2
                left, right = xs[:mid], xs[mid:]
                if left == right:
                    ans += x
    return ans
            
```

```python
part1('input/day2_test.txt')
```

```python
part1('input/day2.txt')
```

```python
def find_factors(n):
    factors = [1]
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            factors.append(i)
            if i != n // i:
                factors.append(n // i)
    return factors


def part2(input_file):
    ranges = parse(input_file)
    ans = 0
    for low, up in ranges:
        for x in range(low, up + 1):
            if x < 10:
                continue
            xs = str(x)
            factors = find_factors(len(xs))
            for f in factors:
                left = xs[:f]
                if left * (len(xs) // f) == xs:
                    # print(x)
                    ans += x
                    break
    return ans
```

```python
part2('input/day2_test.txt')
```
