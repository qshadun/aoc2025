```python
from functools import cache
from collections import defaultdict

def parse_input(input_file):
    graph = defaultdict(list)
    with open(input_file) as f:
        for line in f:
            a, bs = line.strip().split(': ')
            bs = bs.split()
            graph[a].extend(bs)
    return graph


def part1(input_file):
    graph = parse_input(input_file)
    
    @cache
    def dp(cur):
        if cur == 'out':
            return 1
        ans = 0
        for nxt in graph[cur]:
            ans += dp(nxt)
        return ans
    return dp('you')


def part2(input_file):
    graph = parse_input(input_file)
    
    @cache
    def dp(cur, dac, fft):
        if cur == 'out':
            if dac and fft:
                return 1
            else:
                return 0
        ans = 0
        if cur == 'dac':
            dac = True
        elif cur == 'fft':
            fft = True
        for nxt in graph[cur]:
            ans += dp(nxt, dac, fft)
        return ans
    return dp('svr', False, False)
```
