```python
def parse_input(input_file):
    grid = []
    with open(input_file) as f:
        for line in f:
            grid.append(list(line.strip()))
    return grid

def part1(input_file):
    ans = 0
    grid = parse_input(input_file)
    m, n = len(grid), len(grid[0])
    for i in range(1, m):
        for j in range(n):
            if grid[i-1][j] == 'S':
                if grid[i][j] == '^':
                    ans += 1
                    if j > 0 and grid[i][j-1] == '.':
                        grid[i][j-1] = 'S'
                    if j < n-1 and grid[i][j+1] == '.':
                        grid[i][j+1] = 'S'
                else:
                    grid[i][j] = 'S'
    return ans


def part2(input_file):
    
    grid = parse_input(input_file)
    m, n = len(grid), len(grid[0])
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 'S':
                grid[i][j] = 1
            elif grid[i][j] == '.':
                grid[i][j] = 0
    # print(grid)
    for i in range(1, m):
        for j in range(n):
            if grid[i-1][j] != 0 and grid[i-1][j] != '^':
                if grid[i][j] == '^':
                    if j > 0:
                        grid[i][j-1] += grid[i-1][j]
                    if j < n-1 :
                        grid[i][j+1] += grid[i-1][j]
                else:
                    grid[i][j] += grid[i-1][j]
    # for row in grid:
    #     print(row)
    return sum(x for x in grid[-1] if x != '^')
```
